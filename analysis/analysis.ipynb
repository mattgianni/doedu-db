```python
import pandas as pd
import seaborn as sns
```

```python
merged_df = pd.read_feather('merged.feather')
demo_labels_df = pd.read_feather('demo_labels.feather')
demo_labels_df["demo_str"] = demo_labels_df["demographic_num"].astype(str)
```

```python
corrs = merged_df.corr()["pct_std_met_and_above_math"].sort_values(ascending=False).to_frame()
```

```python
corrs_report = corrs.merge(demo_labels_df[["demo_str", "demographic_name", "student_group"]], left_index=True, right_on='demo_str', how='left').dropna().reset_index(drop=True)
corrs_report
```
